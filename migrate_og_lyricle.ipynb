```python
with open("test_data/song.txt") as f:
    data = f.read()
```

```python
import re
pattern = re.compile(r"song: [\"'](.*) ─ (.*)[\"']")

artist_song_pair = re.findall(pattern, data)
```

```python
import json
from lyricle import Lyricle
from model import SongData
from spotify_api_client import SpotifyAPIClient
from utils import LyricleJSONSerializer

song_data_list = []

for i, (artist, song) in enumerate(artist_song_pair):
    print(i, "out of", len(artist_song_pair))
    try:
        track = SpotifyAPIClient.search_track(song, artist)
        song_data = Lyricle.generate_lyricle(track["title"], track["artist"])
    except:
        song_data = SongData(song, artist, "Song doesn't exist", "Song doesn't exist", "Song doesn't exist")
    
    with open("out/all_validGuesses.json", "r") as f:
        initial_valid_guesses = json.load(f)
    LyricleJSONSerializer.generate_song_list_from_data
    
    song_data_list.append(song_data)
```

```python
song_list = LyricleJSONSerializer.generate_song_list_from_data(song_data_list)
valid_guesses = LyricleJSONSerializer.append_valid_guesses(
        initial_valid_guesses,
        song_list,
    )
LyricleJSONSerializer.to_json_file(valid_guesses, f"all_validGuesses.json")
```

```python
with open("out/main.json", "w") as f:
    json.dump(song_list, f, ensure_ascii=False)
```
