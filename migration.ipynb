```python
import json
```

```python
with open('validGuesses.json') as f:
    current_list = json.loads(f.read())
```

```python
sliced_out = current_list["collection"][:108]
```

```python
lyricle_lst = []
for i, slice in enumerate(sliced_out):
    lyricle = {}
    lyricle['id'] = i + 1
    (artist, title) = slice["song"].split("─")
    lyricle['title'] = title.strip()
    lyricle['artist'] = artist.strip()
    lyricle['lyrics'] = slice["lyrics"]
    lyricle['songLink'] = slice['soundcloudLink'] if 'soundcloudLink' in slice else ""
    lyricle['artworkLink'] = slice['artworkLink'] if 'artworkLink' in slice else ""
    lyricle_lst.append(lyricle)
```

```python
out = json.dumps(lyricle_lst, ensure_ascii=False)
with open("out/lyricle.json", 'w') as f:
    f.write(out)
```

```python
artists = {}
for i, slice in enumerate(sliced_out):
    (artist, title) = slice["song"].split("─")
    artist = artist.strip()
    title = title.strip()
    if artist in artists:
        artists[artist].append(title.strip())
    else:
        artists[artist.strip()] = [title]
```

```python
valid_guesses = []
for artist in artists:
    obj = {
        "artist": artist,
        "songs": artists[artist]
    }
    valid_guesses.append(obj)
```

```python
out = json.dumps({
    "collection": valid_guesses
})
with open("out/validGuesses.json", 'w') as f:
    f.write(out)
```

```python
import json
lartist = "hindi"
with open(f'out/{lartist}_choices.json') as f:
    eighties = json.loads(f.read())
    
```

```python
lyricle_lst = []
for i, slice in enumerate(eighties):
    lyricle = {}
    lyricle['id'] = i + 1
    lyricle['title'] = slice["data"]['title']
    lyricle['artist'] = slice["data"]['artist']
    lyricle['lyrics'] = slice["data"]['lyrics']
    lyricle['songLink'] = slice["data"]['songLink']
    lyricle['artworkLink'] = slice["data"]['artworkLink']
    lyricle_lst.append(lyricle)
```

```python
out = json.dumps(
     lyricle_lst
, ensure_ascii=False)
with open(f"out/{lartist}.json", 'w') as f:
    f.write(out)
```

```python
with open(f"out/{lartist}_validGuesses.json", 'w') as f:
    f.write('[]')
```

```python

with open(f'out/{lartist}_validGuesses.json') as f:
    valid_guesses = json.loads(f.read())
catalog = set()
for guess in valid_guesses:
    for title in guess["songs"]:
        catalog.add((title, guess["artist"]))
        
for lyricle in lyricle_lst:
    title = lyricle["title"]
    artist = lyricle["artist"]
    if (title, artist) not in catalog:
        found = False
        for i, search in enumerate(valid_guesses):
            if search['artist'] == artist:
                found = True
                valid_guesses[i]['songs'].append(title)
                break
        if not found:
            valid_guesses.append({"artist": artist, "songs": [title]})
        catalog.add((title, artist))
```

```python
out = json.dumps(valid_guesses, ensure_ascii=False)
with open(f"out/{lartist}_validGuesses.json", 'w') as f:
    f.write(out)
```

```python
with open('out/10s_choices.json') as f:
    seventies = f.read()
    seventies = json.loads(seventies)
```

```python
new_List = []

for i, seventy in enumerate(seventies):
    new = {}
    seventy.pop("song_id")
    new['id'] = i+1
    new.update(seventy['data'])
    new_List.append(new)
```

```python
out = json.dumps(new_List, ensure_ascii=False)
with open("out/10s.json", 'w') as f:
    f.write(out)
```
